# CAEVI

```python
import sqlite3
con = sqlite3.connect("caevi_database.db")
cur = con.cursor()
res = cur.execute("SELECT * FROM caevi")
datos = res.fetchall()
```

```python
column_names = [description[0] for description in cur.description]
column_names
```

```python
from utils.config import Config
from utils.database import get_mongo_client_debug

config = Config()

client = get_mongo_client_debug(config)
caevi_db = client.caevi
caevi_collection = caevi_db.registry

for registry in datos:
    info_dict = {}
    for i, column in enumerate(column_names):
        info_dict[column] = registry[i]
    caevi_collection.insert_one(info_dict)
```
