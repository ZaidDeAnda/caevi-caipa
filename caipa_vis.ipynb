```python
from utils.config import Config
from utils.database import get_mongo_client_debug
from utils.database import create_dataframe_from_cursor

config = Config()

client = get_mongo_client_debug(config)
caipa_db = client.caipa
caipa_collection = caipa_db.registry
caipa_cursor = caipa_collection.find({})
caipa_df = create_dataframe_from_cursor(caipa_cursor)
caipa_df = caipa_df.drop("id", axis=1)
caipa_df = caipa_df.set_index(caipa_df["centro"])
caipa_df = caipa_df.drop("centro", axis=1)
caipa_df
```

```python
import pandas as pd

from CAIPA.utils.dict_utils import val_servicios

valoracion_keys = list(val_servicios.keys())

valoracion_keys.remove("jovenes_graduados")

val_servicios_df = caipa_df[valoracion_keys]
val_servicios_df = val_servicios_df.replace("No aplica", 0)
```

```python
valoracion_categorical_df = pd.DataFrame(columns = ["centro", "servicio", "cantidad"])

columns = list(val_servicios_df.columns)

for row in val_servicios_df.iterrows():
    centro = row[0]
    for col in columns:
        valoracion_categorical_df = valoracion_categorical_df.append(
            {
                "centro" : centro,
                "servicio" : col,
                "cantidad" : int(row[1][col])
            },
            ignore_index=True
        )
```

```python
valoracion_categorical_df
```

```python
promedio = val_servicios_df.groupby("centro").sum().describe()
promedio
```

```python
import plotly.express as px

px.histogram(valoracion_categorical_df, x="centro", y="cantidad", color="servicio")
```
